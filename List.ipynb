This is list. It is ordered, changeable and allow duplicates. Ordered means every item has index. We can add or remove items to list. Since it is indexed it allows duplicates. IN PYTHON LISTS ARE DEFINED AS TYPE LIST

```python
fruits_list = ["Apple","Orange","Banana"]
for x in fruits_list:
    print(x)
fruits_list.append("Mango")
print(fruits_list)
print(fruits_list[0])
num_list = [1,2,3,4,5]
print(num_list)
mixed_list = ["Apple",2,"Orange",3,"banana",4]
print (mixed_list)
print(type(mixed_list))
print(len(mixed_list))
print(type(mixed_list[0]))
print(type(mixed_list[1]))
print(fruits_list[-1])
print(fruits_list[-2])
print(mixed_list[2:5])
```

Changing values in List.The insert() method inserts an item at the specified index. To add an item to the end of the list, use the append() method.The extend() method does not have to append lists, you can add any iterable object (tuples, sets, dictionaries etc.).



```python
fruits_list[1] = "Kiwi"
print(fruits_list)
fruits_list.insert(2,"Watermelon")
print(fruits_list)
#fruits_list.extend(num_list)
new_list = []
new_list.extend(num_list)
print(new_list)
new_list.extend(mixed_list)
print(new_list)
print(fruits_list)
```

The remove() method removes the specified item. Note that this is case sensitive. in beelow case "mango" dont work.
The pop() method removes the specified index.

```python
fruits_list.remove("Mango")
print(fruits_list)
num_list.pop(2)
print(num_list)
```

Looping through LIST

```python
thislist = ["apple", "banana", "cherry","Kiwi"]
for x in thislist:
  print(x)
thislist1 = ["apple", "banana", "cherry"]
i = 0
print("using while loop")
while i < len(thislist1):
  print(thislist1[i])
  i = i + 1
print("using range")
for i in range(len(thislist)):
  print(thislist[i])
```
