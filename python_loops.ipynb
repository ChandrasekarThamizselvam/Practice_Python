```python
start_num = int(input("Enter Starting Number: "))
end_num = int(input("Enter Ending Number: "))
print(f"{start_num}is the Starting Number and {end_num} is the Ending Number")
for(i) in range(start_num,end_num + 1):
    print(i)
```

```python
for i in range(start_num,end_num + 1,3):
    print(i)
```

```python
for i in range(end_num,start_num,-1):
    print(i)
```
